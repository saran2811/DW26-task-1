```python
import re
import getpass
class registration:
    def __init__(self,userid,password):
        self.userid=userid
        self.p=password
    def checkid(self):
         email_cond='[A-Za-z0-9_+*]+@[A-Za-z0-9]+\.[A-Za-z]{2,}'
         if re.search("^[0-9]",self.userid):
            print("invalid mail")
         else:
             if(re.fullmatch(email_cond,self.userid)):
                print("valid mailid")
                file = open('Credentials','a')
                file.writelines(f'{self.userid}')
                file.close()
             else:
                print("invalid mail")
    def checkpwd(self):
        p=self.p
        flag=1
        while True:
            if(len(p)<5 and len(p)>16):
                flag=0
                break
            elif not re.search("[a-z]",p):
                flag=0
                break
            elif not re.search("[A-Z]",p):
                flag=0
                break
            elif not re.search("[0-9]",p):
                flag=0
                break
            elif not re.search("[@$#*]",p):
               flag=0
               break
            elif re.search("\s",p):
               flag=0
               break
            else:
               print("valid password")
               print('Account registration completed\n')
               file = open('Credentials','a')
               pwd=self.p
               file.writelines(f"\t{pwd}\n")
               file.close()
               break
        if flag==0:
            print("invalid password")
            print('Please read the instruction and enter in correct format\n')
    def secquestion(self,question,answer):
        self.question=question
        self.ans=answer
        file=open('security','a')
        file.writelines(self.userid)
        file.writelines(f'\t{self.question}')
        file.writelines(f'\t{self.ans}')
        file.writelines(f"\t{self.p}\n")
        file.close()
        
class login:
    def __init__(self,username,password):
        self.id=username
        self.pwd=password
    def checklogid(self):
        file=open('Credentials','r') 
        with open(r"Credentials", 'r') as fp:
            i=len(fp.readlines())
        for j in range(i+1):
            a=file.readline()
            if self.id in a:
                print('valid ID Entered')
                break
            else:
                pass
            if j==i:
                print('Account doesn\'t exist please register\n')    
    def code(self):
        
        file=open('Credentials','r') 
        with open(r"Credentials", 'r') as fp:
            i=len(fp.readlines())
        for j in range(i+1):
            a=file.readline()
            if self.id in a:
                if self.pwd in a:
                    print('valid password entered')
                    print('login successful\n')
                    break
                else:
                    print('enter a valid password or click forgot password\n')
            else:
                pass
            if j==i:
                print('Account doesn\'t exist please register\n')
class recover:
    def __init__(self,email):
        self.mail=email
        file1=open('security','r')
        with open(r"security", 'r') as fp:
            i=len(fp.readlines())
        for j in range(i+1):
            a=file1.readline()
            if self.mail in a:
                file1=open('security','r')
                txt=file1.readlines()
                A=txt[j].split()  
                b=' '.join(A[1:(len(A)-2)])
                print(b)
                c=input('Your answer : ')
                if c in a:
                   print('Your password is : ',A[(len(A)-1)])
                   break
                else:
                   print('Enter correct answer')
            else:
                pass
            if j==i:
                print('Credential not found please register\n')

while True:
    print('1-New User Registration: \n2-Login: \n3-Forgot password')
    i=int(input('CHOOSE : '))
    if i==1:
        print('Enter emailid :')
        print('Email id should not start with number,shall contain \'*,_,+ \' special characters')
        email=input()
        print('Enter password : ')
        print('1-Password should be of length 5 to maximum 16 characters \n2-Password should contain atleast one Uppercase letter \n3-Password should contain atleast one lowercase letter \n4-Should contain any of following special character\'@,$,#,*\' \n5-Should contain atleast a number \n5-Should not contain any spaces')
        p=getpass.getpass()
        print('Enter the security question of your choice :')
        q=input()
        print('Enter answer in a single word :')
        ans=input()
        user=registration(email,p)
        user.checkid()
        user.checkpwd()
        user.secquestion(q,ans)
    if i==2:
        print('Enter emailid :')
        email=input()
        print('Enter password : ')
        p=getpass.getpass()
        log=login(email,p)
        log.checklogid()
        log.code()
    if i==3:
        print('Enter emailid :')
        email=input()
        log=recover(email)
                    
    


                
                
 
         
```
